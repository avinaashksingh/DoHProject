<h1>Graphify</h1>
<h3>This notebook is going to read the CSV files inputed and convert them into Graphs representing<br>
    a) Time taken in every iteration of requesting a particular domain from the server, and <br>
    b) Number of times the time taken was in a particular range.</h3>

<h4>A python function "graphify" that takes the filepath as input and outputs the two graphs.</h4>
Two types of graphs are needed:<br>
A graph that can simply plot value of time taken on the y axis(max-min) and the iteration number on the x axis(1-100).<br>
A graph that can represent the frequency of value of time taken in a specific range.<br>

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
def dclip(data):
    dmin = np.percentile(data, 25)
    dmax = np.percentile(data, 75)
    data = np.clip(data, dmin, dmax)
    return data

    
```

```python
def graphify_box(domain):
    data_doh_Cloudflare = pd.read_csv("CSVs/{}+doh+Cloudflare.csv".format(domain), usecols = ["Time"])
    data_dns_Cloudflare = pd.read_csv("CSVs/{}+dns+Cloudflare.csv".format(domain), usecols = ["Time"])
    data_doh_Google = pd.read_csv("CSVs/{}+doh+Google.csv".format(domain), usecols = ["Time"])
    data_dns_Google = pd.read_csv("CSVs/{}+dns+Google.csv".format(domain), usecols = ["Time"])
    data_doh_Local = pd.read_csv("CSVs/{}+doh+Local.csv".format(domain), usecols = ["Time"])
    data_dns_Local = pd.read_csv("CSVs/{}+dns+Local.csv".format(domain), usecols = ["Time"])
    data_doh_OpenDNS = pd.read_csv("CSVs/{}+doh+Local.csv".format(domain), usecols = ["Time"])
    data_dns_OpenDNS = pd.read_csv("CSVs/{}+dns+Local.csv".format(domain), usecols = ["Time"])
    data_doh_Quad9 = pd.read_csv("CSVs/{}+doh+Quad9.csv".format(domain), usecols = ["Time"])
    data_dns_Quad9 = pd.read_csv("CSVs/{}+dns+Quad9.csv".format(domain), usecols = ["Time"])
    data_doh_Adguard = pd.read_csv("CSVs/{}+doh+Adguard.csv".format(domain), usecols = ["Time"])
    data_dns_Adguard = pd.read_csv("CSVs/{}+dns+Adguard.csv".format(domain), usecols = ["Time"])
    x = np.arange(1, 101)
 
    y_doh_Cloudflare = data_doh_Cloudflare["Time"]
    y_dns_Cloudflare = data_dns_Cloudflare["Time"]
    y_doh_Google = data_doh_Google["Time"]
    y_dns_Google = data_dns_Google["Time"]
    y_doh_Local = data_doh_Local["Time"]
    y_dns_Local = data_dns_Local["Time"]
    y_doh_OpenDNS = data_doh_OpenDNS["Time"]
    y_dns_OpenDNS = data_dns_OpenDNS["Time"]
    y_doh_Quad9 = data_doh_Quad9["Time"]
    y_dns_Quad9 = data_dns_Quad9["Time"]
    y_doh_Adguard = data_doh_Adguard["Time"]
    y_dns_Adguard = data_dns_Adguard["Time"]

    data = [y_doh_Cloudflare, y_dns_Cloudflare, y_doh_Google, y_dns_Google, y_doh_Local, y_dns_Local,
            y_doh_OpenDNS, y_dns_OpenDNS, y_doh_Quad9, y_dns_Quad9, y_doh_Adguard, y_dns_Adguard]
    data_xt = ["Cloudflare DoH", "Cloudflare DNS", "Google DoH", "Google DNS", "Local DoH", "Local DNS",
               "OpenDNS DoH", "OpenDNS DNS", "Quad9 DoH", "Quad9 DNS", "Adguard DoH", "Adguard DNS"]
    fig = plt.figure(figsize=(18,18))
    
    ax = fig.add_subplot(1,1,1) 

    ax.boxplot(data, showfliers=False)

    ax.set_title('"{}"'.format(domain), fontsize=40)
    ax.set_ylabel("Time Taken", fontsize=40)
    ax.set_xticklabels(data_xt, rotation=45, fontsize=20)
    ax.grid()
    
    
    #plt.ylim(0, dns.max() + dns.mean())
    fig.savefig("Boxplots/{}+box.png".format(domain))
    plt.show()
```

```python

def graphify_plot(domain):
    data_doh_Cloudflare = pd.read_csv("CSVs/{}+doh+Cloudflare.csv".format(domain), usecols = ["Time"])
    data_dns_Cloudflare = pd.read_csv("CSVs/{}+dns+Cloudflare.csv".format(domain), usecols = ["Time"])
    data_doh_Google = pd.read_csv("CSVs/{}+doh+Google.csv".format(domain), usecols = ["Time"])
    data_dns_Google = pd.read_csv("CSVs/{}+dns+Google.csv".format(domain), usecols = ["Time"])
    data_doh_Local = pd.read_csv("CSVs/{}+doh+Local.csv".format(domain), usecols = ["Time"])
    data_dns_Local = pd.read_csv("CSVs/{}+dns+Local.csv".format(domain), usecols = ["Time"])
    data_doh_OpenDNS = pd.read_csv("CSVs/{}+doh+Local.csv".format(domain), usecols = ["Time"])
    data_dns_OpenDNS = pd.read_csv("CSVs/{}+dns+Local.csv".format(domain), usecols = ["Time"])
    data_doh_Quad9 = pd.read_csv("CSVs/{}+doh+Quad9.csv".format(domain), usecols = ["Time"])
    data_dns_Quad9 = pd.read_csv("CSVs/{}+dns+Quad9.csv".format(domain), usecols = ["Time"])
    data_doh_Adguard = pd.read_csv("CSVs/{}+doh+Adguard.csv".format(domain), usecols = ["Time"])
    data_dns_Adguard = pd.read_csv("CSVs/{}+dns+Adguard.csv".format(domain), usecols = ["Time"])
    x = np.arange(1, 101)
 
    y_doh_Cloudflare = dclip(data_doh_Cloudflare["Time"])
    y_dns_Cloudflare = dclip(data_dns_Cloudflare["Time"])
    y_doh_Google = dclip(data_doh_Google["Time"])
    y_dns_Google = dclip(data_dns_Google["Time"])
    y_doh_Local = dclip(data_doh_Local["Time"])
    y_dns_Local = dclip(data_dns_Local["Time"])
    y_doh_OpenDNS = dclip(data_doh_OpenDNS["Time"])
    y_dns_OpenDNS = dclip(data_dns_OpenDNS["Time"])
    y_doh_Quad9 = dclip(data_doh_Quad9["Time"])
    y_dns_Quad9 = dclip(data_dns_Quad9["Time"])
    y_doh_Adguard = dclip(data_doh_Adguard["Time"])
    y_dns_Adguard = dclip(data_dns_Adguard["Time"])
    
    
    
    fig = plt.figure(figsize=(18,18))
    
    ax = fig.add_subplot(1,1,1) 

    ax.plot(x, y_dns_Cloudflare, color="red", label="Cloudflare DNS")
    ax.plot(x, y_doh_Cloudflare, color="red", linestyle=":", label="Cloudflare DoH")
    ax.plot(x, y_dns_Google, color="green", label="Google DNS")
    ax.plot(x, y_doh_Google, color="green", linestyle=":",label="Google DoH")
    ax.plot(x, y_dns_Local, color="magenta", label="Local DNS")
    ax.plot(x, y_doh_Local, color="magenta", linestyle=":",label="Local DoH")
    ax.plot(x, y_dns_OpenDNS, color="cyan", label="OpenDNS DNS")
    ax.plot(x, y_doh_OpenDNS, color="cyan", linestyle=":",label="OpenDNS DoH")
    ax.plot(x, y_dns_Quad9, color="yellow", label="Quad9 DNS")
    ax.plot(x, y_doh_Quad9, color="yellow", linestyle=":",label="Quad9 DoH")
    ax.plot(x, y_dns_Adguard, color="blue", label="Adguard DNS")
    ax.plot(x, y_doh_Adguard, color="blue", linestyle=":",label="Adguard DoH")

    ax.set_xlabel("Iterations", fontsize=20)
    fig.suptitle('"{}"'.format(domain), fontsize=40)
    ax.set_ylabel("Time Taken", fontsize=20)
    
    ax.grid()
    ax.legend(prop={'size': 20})
    fig.savefig("Simpleplots/{}.png".format(domain))
    plt.show()
    
```

```python
domains = ["gla.ac.uk",
          "nationalgeographic.co.uk",
          "facebook.com",
          "google.com",
          "kjhdfce.com",
          "kjhdfce.co.uk",
          "cbse.nic.in",
          "the_omnia.com",
          "penguin.com",
          "Aynrand.com"]
resolvers = ["Cloudflare",
             "Google",
             "Local",
             "OpenDNS",
             "Quad9",
             "Adguard"]
```

```python
for domain in domains:
    graphify_plot(domain)
```

```python
for domain in domains:
    graphify_box(domain)
```
