<h1>Graphify</h1>
<h3>This notebook is going to read the CSV files inputed and convert them into Graphs representing<br>
    a) Time taken in every iteration of requesting a particular domain from the server, and <br>
    b) Number of times the time taken was in a particular range.</h3>

<h4>A python function "graphify" that takes the filepath as input and outputs the two graphs.</h4>
Two types of graphs are needed:<br>
A graph that can simply plot value of time taken on the y axis(max-min) and the iteration number on the x axis(1-100).<br>
A graph that can represent the frequency of value of time taken in a specific range.<br>

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def graphify(domain, resolver):
    data_doh = pd.read_csv("{}+doh+{}.csv".format(domain, resolver), usecols = ["Time"])
    data_dns = pd.read_csv("{}+dns+{}.csv".format(domain, resolver), usecols = ["Time"])
    x = np.arange(1, 101)
 
    y_doh = data_doh["Time"]
    y_dns = data_dns["Time"]

    
    fig = plt.figure(figsize=(18,18))
    
    ax = fig.add_subplot(1,1,1) 

    ax.scatter(x, y_dns, color="red", s=40, marker = "v", label="DNS")
    ax.scatter(x, y_doh, color="blue", s=40, marker = "v", label="DoH")
    ax.scatter([50], y_dns.mean(), color='red', marker="s", s=100, label="Mean DNS")
    ax.scatter([50], y_doh.mean(), color='blue',marker="s", s=100, label="Mean DoH")

    ax.set_xlabel("Iterations", fontsize=20)
    fig.suptitle('"{}" requested from {}'.format(domain, resolver), fontsize=20)
    ax.set_ylabel("Time Taken", fontsize=20)
    
    ax.grid()
    ax.legend()
    plt.ylim(0, 0.2)
    fig.savefig("{}+{}.png".format(domain, resolver))
    plt.show()
    
```

```python
domains = ["gla.ac.uk",
          "nationalgeographic.co.uk",
          "facebook.com",
          "google.com",
          "kjhdfce.com",
          "kjhdfce.co.uk",
          "cbse.nic.in",
          "the_omnia.com",
          "penguin.com",
          "Aynrand.com"]
resolvers = ["Cloudflare",
             "Google",
             "Local",
             "OpenDNS",
             "Quad9",
             "Adguard"]
```

```python
for domain in domains:
    for resolver in resolvers:
        graphify(domain, resolver)
```
